```python
import matplotlib.pyplot as plt

class Titik:
    def __init__(self, x, y, nama=""):
        self.x = x
        self.y = y
        self.nama = nama

    def cermin_sumbu_x(self):
        return Titik(self.x, -self.y, self.nama + "'")

    def to_tuple(self):
        return (self.x, self.y)


class Bangun:
    def __init__(self, titik_list):
        self.titik = titik_list

    def cermin_sumbu_x(self):
        return Bangun([t.cermin_sumbu_x() for t in self.titik])

    def get_koordinat(self):
        x = [t.x for t in self.titik] + [self.titik[0].x]
        y = [t.y for t in self.titik] + [self.titik[0].y]
        return x, y

    def label_titik(self):
        for t in self.titik:
            plt.text(t.x, t.y, t.nama)


A = Titik(2, 5, "A")
B = Titik(3, 2, "B")
C = Titik(4, 3, "C")
D = Titik(2, 2, "D")

bangun_asli = Bangun([A, B, C, D])
bangun_cermin = bangun_asli.cermin_sumbu_x()

plt.figure(figsize=(6, 8))

x1, y1 = bangun_asli.get_koordinat()
plt.plot(x1, y1, marker='o')
plt.fill(x1, y1, alpha=0.3)
bangun_asli.label_titik()

x2, y2 = bangun_cermin.get_koordinat()
plt.plot(x2, y2, marker='o')
plt.fill(x2, y2, alpha=0.3)
bangun_cermin.label_titik()

plt.axhline(0)
plt.axvline(0)

plt.title("Refleksi terhadap Sumbu X")
plt.grid(True)
plt.axis("equal")

plt.show()
```
